import libraries

```python
import torch
from transformers import pipeline
import pandas as pd
```

Define Candidate Tags

```python
labels = ["Billing", "Technical Issue", "Account Management", "Login Problem"]
```

Create a Sample Dataset

```python
data = {
    "Ticket ID": [1, 2, 3, 4],
    "Text": [
        "My internet is not working since yesterday.",
        "I want to change my billing address.",
        "Getting login error on the mobile app.",
        "How to reset my password?"
    ]
}

df = pd.DataFrame(data)
df
```

Load the Zero-Shot Classification Model

```python
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

Classify Each Ticket and Predict Tags

```python
def classify_ticket(text):
    result = classifier(text, candidate_labels=labels)
    return result["labels"][:3]  # Top 3 predicted tags

# Apply to all tickets
df["Top Tags"] = df["Text"].apply(classify_ticket)
df
```
